**The business team will provide vouchers for customers who are indicated to have the potential to churn next month. The following are the criteria for churn next month:**

1. recency > 60
2. frequent < 2
3. monetary < 125000

```python
import pandas as pd
customer = {'user_id': ['user311', 'user576'],
            'recency': [34, 78],
            'frequent': [1,0],
            'monetary': [85000, 0]}

customer = pd.DataFrame(customer)
customer
```

```python
potential_list = []

for index, column in customer.iterrows():
    print('\nUser ID:', column['user_id'])
    if column['recency'] > 60:
        criteria_recency = 'Fulfilled'
        if column['frequent'] < 2:
            criteria_frequency = 'Fulfilled'
            if column['monetary'] < 125000:
                criteria_monetary = 'Fulfilled'
                churn = 'Yes'
            else:
                criteria_monetary = 'Not Fulfilled'
                churn = 'No'
        else:
            criteria_frequency = 'Not Fulfilled'
            churn = 'No'

    else:
        criteria_recency = 'Not Fulfilled'
        churn = 'No'
        if column['frequent'] < 2:
            criteria_frequency = 'Fulfilled'
            if column['monetary'] < 125000:
                criteria_monetary = 'Fulfilled'
                
            else:
                criteria_monetary = 'Not Fulfilled'
                
        else:
            criteria_frequency = 'Not Fulfilled'
            

    potential_list.append(churn)
    print('Criteria Recency:', criteria_recency)
    print('Criteria Frequency:', criteria_frequency)
    print('Criteria Monetary:', criteria_monetary)
    

customer['potential_churn'] = potential_list
customer
```

**A series of numbers between 1 - 500 which are multiples of 5 which are divisible by 4**

```python
bil = []
n_bil = 0
for i in range(1,501):
    if i%5 == 0:
        if i%4 == 0:
            bil.append(i)
            n_bil +=1

print('Number Series:', bil)
print('Total Numbers:', n_bil)
```

**The company's security team will create a special team (containing 2 people per team) to guard every night during the New Year holiday. The company has 8 members of the security unit. Determine how many possible teams can be formed.**

```python
list_id_member = ['a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7' ,'a8']

from itertools import combinations
team = combinations(list_id_member, 2)

prob_team = 0
for i in list(team):
    prob_team += 1
    print(i)

print('\nPossible teams that can be formed:', prob_team)
```

**The customer loyalty team wants to give points to customer to retain their loyalty. The criteria of customer are:**

1. Customer in java with total order less than or equal to 100 will be granted 1 point, while total order more than 100 will be granted 3 point
2. Customer other than in java with total order less than or equal to 50 will be granted 1 point, while total order more than 50 will be granted 3 point

**Determine how many percent of users get additional 3 points for each region (Java & Not Java).**

```python
import pandas as pd
data = pd.read_csv('exercise_programming1.csv', sep=',')
data.head()
```

```python
point_list = []

cust_java = 0
cust_non_java = 0

java_point_3 = 0
non_java_point_3 = 0

for index, column in data.iterrows():
    if column['is_java'] == True:
        cust_java += 1
        if column['order'] > 100:
            point = 3
            java_point_3 += 1
        else:
            point = 1
    else:
        cust_non_java += 1
        if column['order'] > 50:
            point = 3
            non_java_point_3 += 1
        else:
            point = 1
    
    point_list.append(point)
    
java_region = round(java_point_3/cust_java*100, 2)
non_java_region = round(non_java_point_3/cust_non_java*100, 2)

data['additional_point'] = point_list
print('Percentage Customer Region Java:', java_region)
print('Percentage Customer Region Non Java:', non_java_region)
data
```
